```python
all_types = {"rap", "rnb", "pop", "heavy metal", "classical", "indie", "jazz", "rock", "blues", "edm"}
group_1 = {"rap", "rnb", "pop", "jazz", "rock"}
group_2 = {"blues", "rap", "rock", "edm"}

# Definitely play (common in both groups)
definitely_play = group_1.intersection(group_2)

# Perhaps play (only in one group)
perhaps_play = group_1.symmetric_difference(group_2)

# Do not play (not in either group)
do_not_play = all_types.difference(group_1.union(group_2))

print("Definitely play:", definitely_play)
print("Perhaps play:", perhaps_play)
print("Do not play:", do_not_play)
```
